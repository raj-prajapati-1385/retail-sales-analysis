#(3) install / import libraries

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#(4) Clean the data ( Data Cleaning )

```python
# Load & inspect data

df = pd.read_csv("retail_price.csv")

df.head()   # See first 5 rows

df.tail() # See last 5 rows

df.info()  # Data types, missing values

df.describe() # Stats summary for numerical columns

df.shape # Rows and columns count
```

```python
# This code checks for missing values in each column of
# your DataFrame df and then sums them up

df.isnull().sum()
```

- There is no missing values

```python
# remove duplicates

df.duplicated().sum()
df = df.drop_duplicates()
```

```python
#  Correct Data Types
# Dates should be datetime, numbers should be int or float.

df['month_year'] = pd.to_datetime(df['month_year'])
df['qty'] = df['qty'].astype(int)
df['unit_price'] = df['unit_price'].astype(float)
```

```python
# Standardize Text Data
# Make product names or categories consistent.

df['product_category_name'] = df['product_category_name'].str.strip().str.title()
```

```python
# Final Check

df.info()
df.head()
```

#(5) Compute KPIs & Insights

```python
total_sales = df['total_price'].sum()
total_orders = len(df)
avg_order_value = total_sales / total_orders
top_products = df.groupby('product_id')['total_price'].sum().sort_values(ascending=False).head(10)

print("Total sales : ",total_sales)
print("Total Orders : ", total_orders)
print("avg. order value : ", avg_order_value)
top_products
```

#(6) Visualization

```python
# Monthly analysis & plots

df['Month'] = df['month_year'].dt.to_period('M')
monthly = df.groupby('Month')['total_price'].sum().reset_index()
monthly['Month'] = monthly['Month'].astype(str)

plt.figure(figsize=(10,5))
sns.lineplot(data=monthly, x='Month', y='total_price', marker='o')
plt.xticks(rotation=45)
plt.title('Monthly Sales')
plt.show()
plt.savefig("my_graph.jpg", dpi=300, bbox_inches='tight')
```

#(7) Final Outputs

```python
# Save outputs (clean CSV and images)

df.to_csv('clean_sales.csv', index=False)
monthly.to_csv('monthly_sales.csv', index=False)
plt.savefig('monthly_sales.png', bbox_inches='tight')
```
